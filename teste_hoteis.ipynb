```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

### Criando a busca dos dados

```python
# Onde e como navegar
url = 'https://www.tripadvisor.com.br/Restaurant_Review-g303506-d983060-Reviews-Restaurante_Marius_Degustare-Rio_de_Janeiro_State_of_Rio_de_Janeiro.html'
headers = {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}

site = requests.get(url, headers = headers)
soup = BeautifulSoup(site.content, 'html.parser')

#selecionar as tags html que serão raspadas para a criação do dataset

caixa_comentarios = soup.find('div', class_= 'listContainer hide-more-mobile')
comentarios = caixa_comentarios.find_all('div', class_ = 'mobile-more')
```

### seleção de conteudo

```python
#nome do restaurante
nome = 'Restaurante Marius Degustare'

#comentário
comentarios_texto = comentarios[6].find('div', class_='prw_rup prw_reviews_text_summary_hsx').get_text()

#data publicacao
pdata = str(comentarios[6].find('span', class_= 'ratingDate'))
data_polida = pdata[32::]
data_publicacao = data_polida[:17]

#data visita
data_visita = str(comentarios[6].find('div', class_= 'prw_rup prw_reviews_stay_date_hsx').get_text())
data_visita = data_visita[15:]

dicionario = {"Restaurante": nome, "Comentario" : comentarios_texto, "Data_Publicacao": data_publicacao, "Data_Visita": data_visita}
```

```python
# Função para criar um dicionario com os dados correspondentes lista de colunas 
def dict_setter(lista_itens,lista_colunas):
    dicionario = {}
    n = len(lista_colunas)

    splited = [lista_itens[i::n] for i in range(n)]

    for i in range(0, 4):
        dicionario[lista_colunas[i]] = splited[i]

    return dicionario
```

```python
#criando laço para pegar todos os comentarios
lista_colunas = ['Restaurante','Comentario', "Data Publicacao", "Data Visita"]


def comment_setter(comentarios, nome):
    lista_itens = []
    for c in range(0,7):

        comentarios_texto = comentarios[c].find('div', class_='prw_rup prw_reviews_text_summary_hsx').get_text().strip()


        pdata = str(comentarios[c].find('span', class_= 'ratingDate'))
        data_polida = pdata[32::]
        data_publicacao = data_polida[:17]

        data_visita = str(comentarios[c].find('div', class_= 'prw_rup prw_reviews_stay_date_hsx').get_text().strip())
        data_visita = data_visita[15:]

        lista_itens.append(nome)
        lista_itens.append(comentarios_texto)
        lista_itens.append(data_publicacao)
        lista_itens.append(data_visita)
        
    return lista_itens
```

```python
itens = comment_setter(comentarios= comentarios, nome= nome)
dis = dict_setter(lista_itens= itens, lista_colunas= lista_colunas)
df = pd.DataFrame(dis)
```

```python
url2 = 'https://www.tripadvisor.com.br/Restaurant_Review-g303506-d1646830-Reviews-Talho_Capixaba-Rio_de_Janeiro_State_of_Rio_de_Janeiro.html'

site = requests.get(url2, headers = headers)
soup = BeautifulSoup(site.content, 'html.parser')
caixa_comentarios2 = soup.find('div', class_= 'listContainer hide-more-mobile')
comentarios2 = caixa_comentarios.find_all('div', class_ = 'mobile-more')


avaliacao =  comment_setter(comentarios = comentarios2,  nome = "Talho Capixaba")
dis2 = dict_setter(lista_itens= avaliacao, lista_colunas= lista_colunas)
df2 = pd.DataFrame(dis2)
```

#### Juntando as bases de dados 

```python
df = pd.merge(df, df2,how= 'outer')
df
```
